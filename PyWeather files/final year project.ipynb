```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import xarray as xr

# Load the NetCDF file
file_path = "C:\\Users\\HexTwo-PC\\Desktop\\Python_Programs\\Temperature\\CRU_maximum_temperature_mon_0.5x0.5_global_2009_v4.03.nc"
data = xr.open_dataset(file_path)

# Subset data for Nigeria
nigeria_data = data.sel(lat=slice(4, 14), lon=slice(3, 15))

# Plot using Cartopy
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'projection': ccrs.PlateCarree()})
ax.coastlines(resolution='10m', color='black', linewidth=0.8)  # High-resolution coastlines
ax.set_extent([3, 15, 4, 14])  # Longitude and latitude bounds for Nigeria

# Plot surface temperature
nigeria_data['tasmax'].isel(time=0).plot(ax=ax, cmap='coolwarm', transform=ccrs.PlateCarree(), robust=True)
plt.title("Surface Temperature in Nigeria")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import xarray as xr

# Load the NetCDF file
file_path = "C:\\Users\\HexTwo-PC\\Desktop\\Python_Programs\\Temperature\\CRU_maximum_temperature_mon_0.5x0.5_global_2009_v4.03.nc"
data = xr.open_dataset(file_path)

# Subset data for Nigeria
nigeria_data = data.sel(lat=slice(4, 14), lon=slice(3, 15))  # Latitude: 4°N–14°N, Longitude: 3°E–15°E

# Plot with higher resolution
plt.figure(figsize=(12, 8), dpi=300)  # Increase DPI for clarity
nigeria_data['tasmax'].isel(time=0).plot(cmap='coolwarm', robust=True)  # Use 'robust' for better color scaling
plt.title("Surface Temperature in Nigeria")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()  # Prevents overlapping labels
plt.show()

# Save the plot as a high-resolution image
output_path = "nigeria_surface_temperature_map.png"
plt.savefig(output_path, dpi=300, bbox_inches='tight')  # bbox_inches ensures no cropping
print(f"High-resolution map saved to {output_path}")
```
